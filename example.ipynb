```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import chi2_contingency
import DimensionalityReduction
```

```python
# import the dataset
diabetes = pd.read_csv('diabetes.csv')
# turning it into a classification task
diabetes['obesse'] = np.where(diabetes['BMI'] > 30, 'Yes', 'No')
```

```python
dr = DimensionalityReduction()
# tell DimensionalityReduction what dataset to use(diabetes), what's the target 
dr.recognize(diabetes, 'Outcome', 'obesse')
```

```python
dr.reduce(show=True, plot=True)
```
